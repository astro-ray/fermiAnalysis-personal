```python
sed = gta.sed(ID)
```

```python
c = np.load('OKS1441+25_low_re.npy', allow_pickle=True).flat[0]
```

```python
E = np.array(c['sources'][ID]['model_flux']['energies'])
dnde = np.array(c['sources'][ID]['model_flux']['dnde'])
dnde_hi = np.array(c['sources'][ID]['model_flux']['dnde_hi'])
dnde_lo = np.array(c['sources'][ID]['model_flux']['dnde_lo'])
```

```python
plt.clf()
plt.loglog(E, (E**2)*dnde, 'k--')
plt.loglog(E, (E**2)*dnde_hi, 'k')
plt.loglog(E, (E**2)*dnde_lo, 'k')
plt.errorbar(np.array(sed['e_ctr']),
             sed['e2dnde'], 
             yerr=sed['e2dnde_err'], fmt ='o')
plt.xlabel('E [MeV]')
plt.ylabel(r'E$^{2}$ dN/dE [MeV cm$^{-2}$ s$^{-1}$]')
plt.show()
```

```python
plt.clf()
plt.loglog(E, (E**2)*dnde, 'k--')
plt.loglog(E, (E**2)*dnde_hi, 'k')
plt.loglog(E, (E**2)*dnde_lo, 'k')
plt.errorbar(sed['e_ctr'][:-10],
             sed['e2dnde'][:-10], 
             yerr=sed['e2dnde_err'][:-10], fmt ='o')
plt.errorbar(np.array(sed['e_ctr'][-10:]),
         sed['e2dnde_ul95'][-10:], yerr=0.2*sed['e2dnde_ul95'][-10:], 
             fmt='o', uplims=True)
plt.xlabel('E [MeV]')
plt.ylabel('E$^{2}$ dN/dE [MeV cm$^{-2}$ s$^{-1}$]')
plt.show()
```
